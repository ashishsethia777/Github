```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# Number of components required to get P% variance  
#pca = PCA(0.96).fit(X_train_scaled)
#print(pca.n_components_)

pca = PCA().fit(X_train_scaled)

#The amount of variance that each PC explains
print(pca.explained_variance_ratio_*100)

#Cumulative Variance explained
var=np.cumsum(np.round(pca.explained_variance_ratio_, decimals=4)*100)
print(var)

plt.plot(var)
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn import tree 
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import AdaBoostClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.multiclass import OneVsOneClassifier


```

```python
from sklearn.pipeline import make_pipeline
from sklearn import preprocessing

base_estimator=tree.DecisionTreeClassifier()
pipeline = make_pipeline(preprocessing.StandardScaler(),
                        OneVsRestClassifier(AdaBoostClassifier(base_estimator=base_estimator)))
#pipeline= OneVsRestClassifier(pipeline)
```

```python
pipeline.get_params()
```

```python
Y_train[:6]
```

```python
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score,average_precision_score,f1_score,recall_score,roc_auc_score
from sklearn.metrics import make_scorer


scoring = {'accuracy': make_scorer(accuracy_score),
            'prec': make_scorer(average_precision_score),
            'f1': make_scorer(f1_score,average='macro'),
            'recall': make_scorer(recall_score,average='macro'),
            'roc_auc': make_scorer(roc_auc_score,average='macro')
          }
             
param_grid = {'onevsrestclassifier__estimator__base_estimator__criterion' : ["gini", "entropy"],
              'onevsrestclassifier__estimator__base_estimator__max_depth' : range(1,25) 
             }


#scoring = ['accuracy', 'precision','roc_auc','f1_weighted']

weights = [1/Y_train.shape[0]] * Y_train.shape[0]
#clf = GridSearchCV(pipeline, param, cv=10, fit_params={'decisiontreeclassifier__sample_weight': weights})

clf = GridSearchCV(estimator=pipeline, param_grid=param_grid,cv=10, 
                       n_jobs=-1, verbose=1,scoring=scoring,refit='prec', return_train_score=True
)   


# Fit and tune model
clf.fit(X_train,Y_train)
```
